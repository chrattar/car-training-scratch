```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('.\\episode_rewards_dfd.csv')
plt.figure(figsize=(12, 6))
plt.plot(df['Episode'], df['Total Reward'])
plt.title('Total Reward per Episode')
plt.grid(True)
plt.xlabel('Episode')
plt.xticks(np.arange(0, df['Episode'].max() + 1, step=max(1, df['Episode'].max() // 10)))
plt.ylabel('Total Reward')
plt.show()
```

```python
window_size = 100
df['Moving Average Reward'] = df['Total Reward'].rolling(window=window_size).mean()

plt.figure(figsize=(12,6))
plt.plot(df['Episode'], df['Moving Average Reward'])
plt.title(f'Moving Average Reward (Window Size: {window_size}')
plt.grid(True)
plt.xlabel('Episode')
plt.xticks(np.arange(0, df['Episode'].max() + 1, step=max(1, df['Episode'].max() // 10)))
plt.ylabel('Average Reward')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Episode'], df['step_count'])
plt.title('Steps per Episode')
plt.grid(True)
plt.xlabel('Episode')
plt.xticks(np.arange(0, df['Episode'].max() + 1, step=max(1, df['Episode'].max() // 10)))
plt.ylabel('Number of Steps')
plt.show()
```
