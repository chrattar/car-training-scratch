```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('.\\episode_rewards_dfd.csv')
plt.figure(figsize=(12, 6))
plt.plot(df['Episode'], df['Total Reward'])
plt.title('Total Reward per Episode')
plt.grid(True)
plt.xlabel('Episode')
plt.xticks(np.arange(0, df['Episode'].max() + 1, step=max(1, df['Episode'].max() // 10)))
plt.ylabel('Total Reward')
plt.show()
```

```python
window_size = 100
df['Moving Average Reward'] = df['Total Reward'].rolling(window=window_size).mean()

plt.figure(figsize=(12,6))
plt.plot(df['Episode'], df['Moving Average Reward'])
plt.title(f'Moving Average Reward (Window Size: {window_size}')
plt.grid(True)
plt.xlabel('Episode')
plt.xticks(np.arange(0, df['Episode'].max() + 1, step=max(1, df['Episode'].max() // 10)))
plt.ylabel('Average Reward')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Episode'], df['step_count'])
plt.title('Steps per Episode')
plt.grid(True)
plt.xlabel('Episode')
plt.xticks(np.arange(0, df['Episode'].max() + 1, step=max(1, df['Episode'].max() // 10)))
plt.ylabel('Number of Steps')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Episode'], df['Epsilon'])
plt.grid(True)
plt.xlabel('Episode')
plt.xticks(np.arange(0, df['Episode'].max() + 1, step=max(1, df['Episode'].max() // 10)))
plt.ylabel('Epsilon')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['step_count'], df['Total Reward'], alpha=0.5)
plt.title('Reward vs. Steps')
plt.xlabel('Number of Steps')
plt.ylabel('Total Reward')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.hexbin(df['Xpos'], df['Ypos'], gridsize=20, cmap='YlOrRd')
plt.title('Car Position Heat Map')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.colorbar(label='Frequency')
plt.show()
```

```python

# Calculate rolling mean and standard deviation
df['Rolling_Mean_Reward'] = df['Total Reward'].rolling(window=window_size).mean()
df['Rolling_Std_Reward'] = df['Total Reward'].rolling(window=window_size).std()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df['Episode'], df['Rolling_Mean_Reward'], label='Rolling Mean')
plt.fill_between(df['Episode'], 
                 df['Rolling_Mean_Reward'] - df['Rolling_Std_Reward'],
                 df['Rolling_Mean_Reward'] + df['Rolling_Std_Reward'],
                 alpha=0.2, label='1 Std Dev')
plt.title(f'Rolling Mean Reward with Standard Deviation (Window: {window_size})')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Total Reward'], bins=50, edgecolor='black')
plt.title('Distribution of Total Rewards')
plt.xlabel('Total Reward')
plt.ylabel('Frequency')
plt.show()
```
